# Introduction
Many financial model frameworks such as Modern Portfolio Theorty and CAPM are underpinned by the assumption that market returns are normally distributed.

As such, time series returns should be tested for normality.

# Mean-Variance Portfolio Theory
When stock returns are normally distributed, optimal portfolios become a function of mean return, variance of return and covariance between different portfolio components.

Mean-variance refers to the mean and variance of the (log) returns of the different securities, which are calculated as follows:

We will select a basket of several stocks.

```python
symbols = ['AAPL','MSFT','DB','GLD']
noa = len(symbols)
```

```python
data = pd.DataFrame()
for sym in symbols:
    data[sym] = web.get_data_yahoo(sym,start='2014-09-12')['Adj Close']
data.columns = symbols
```

Calculate log returns:

```python
rets = np.log(data/data.shift(1))
```

Over the period of the time series, we see significant differences in the annualized performance.  We use a factor of 252 trading days to annualize the daily returns.

```python
rets.mean()*252
```

Calculate covariance matrix of securities:

```python
rets.cov()*252
```

Generate portfolio security weights

```python
weights = np.random.random(noa)
weights /= np.sum(weights)
weights
```

#### The equation for expected portfolio return.

```python
np.sum(rets.mean() * weights) * 252
```

#### The equation for expected portfolio variance.

```python
np.dot(weights.T,np.dot(rets.cov()*252,weights))
```

#### The equation for expected portfolio standard deviation (volatility)

```python
np.sqrt(np.dot(weights.T,np.dot(rets.cov()*252,weights)))
```

### Monte Carlo Simulation of Returns

For a given set of securities, we must discern the risk-return profile.  We will generate random portfolio weighting allocations and record the expected return and variance.

```python
prets = []
pvols = []
for p in range(2500):
    #initialize weights
    weights = np.random.random(noa)
    weights /= np.sum(weights)
    #append returns to list
    prets.append(np.sum(rets.mean()*weights)*252)
    #append volatilities to list
    pvols.append(np.sqrt(np.dot(weights.T,
                               np.dot(rets.cov()*252,weights))))
```

### Sharpe Ratio Plot

The Sharpe Ratio is defined as the expected excess return of a portfolio over the risk-free short rate, divided by the expected standard deviation of the portfolio.

```python
plt.figure(figsize=(8,4))
plt.scatter(pvols,prets,c=np.array(prets)/np.array(pvols),marker='o')
plt.grid(True)
plt.xlabel('expected volatility')
plt.ylabel('expected return')
plt.colorbar(label='Sharpe ratio')
```

#### Define Statistics

```python
def statistics(weights):
    '''Returns portfolio statistics
    Parameters
    ==========
    pret : float
        expected portfolio return
    pvol : float
        expected portfolio volatility
    pret / pvol : float
        Sharpe ratio for rf=0
    '''
    weights = np.array(weights)
    pret = np.sum(rets.mean()*weights)*252
    pvol = np.sqrt(np.dot(weights.T,np.dot(rets.cov()*252,weights)))
    return np.array([pret,pvol,pret/pvol])
```

### Portfolio Optimization

```python
import scipy.optimize as sco
def min_func_sharpe(weights):
    return -statistics(weights)[2]
```

```python
#add constraint that all parameters (weights) add up to 1
cons = ({'type': 'eq', 'fun': lambda x:  np.sum(x) - 1})
#add constraint that each parameter (weights) are within 0 and 1
bnds = tuple((0, 1) for x in range(noa))
#starting parameter list (initial guesses for the weights)
noa * [1. / noa,]
```

```python
%%time
opts = sco.minimize(min_func_sharpe, noa * [1. / noa,], method='SLSQP',
                       bounds=bnds, constraints=cons)
```

```python
opts
```

#### View optimal portfolio

```python
opts['x'].round(3)
```

The following shows the expected return, volatility and sharpe ratio.

```python
statistics(opts['x'].round(3))
```

### Minimize the variance of the portfolio

```python
def min_func_variance(weights):
    return statistics(weights)[1] ** 2
```

```python
optv = sco.minimize(min_func_variance,noa*[1./noa,],
                   method='SLSQP',bounds=bnds,constraints=cons)
```

```python
optv
```

The following portfolio is the absolute minimum variance portfolio.

```python
optv['x'].round(3)
```

The following shows the expected return, volatility and sharpe ratio.

```python
statistics(optv['x'].round(3))
```

### Efficient Frontier

The Efficient Frontier is comprised of all optimal portfolios with a higher return than the absolute minimum variance portfolio.  These portfolios dominate all other portfolios in terms of expected returns given a certain risk level.

We will fix a target return and derive for each such level of portfolio weights that lead to a minimum volatility level.

```python
cons = ({'type': 'eq', 'fun': lambda x:  statistics(x)[0] - tret},
        {'type': 'eq', 'fun': lambda x:  np.sum(x) - 1})
bnds = tuple((0, 1) for x in weights)
```

```python
def min_func_port(weights):
    return statistics(weights)[1]
```

```python
%%time
trets = np.linspace(0.0, 0.25, 50)
tvols = []
for tret in trets:
    cons = ({'type': 'eq', 'fun': lambda x:  statistics(x)[0] - tret},
            {'type': 'eq', 'fun': lambda x:  np.sum(x) - 1})
    res = sco.minimize(min_func_port, noa * [1. / noa,], method='SLSQP',
                       bounds=bnds, constraints=cons)
    tvols.append(res['fun'])
tvols = np.array(tvols)
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(pvols, prets,
            c=np.array(prets) / np.array(pvols), marker='o')
            # random portfolio composition
plt.scatter(tvols, trets,
            c=np.array(trets) / np.array(tvols), marker='x')
            # efficient frontier
plt.plot(statistics(opts['x'])[1], statistics(opts['x'])[0],
         'r*', markersize=15.0)
            # portfolio with highest Sharpe ratio
plt.plot(statistics(optv['x'])[1], statistics(optv['x'])[0],
         'y*', markersize=15.0)
            # minimum variance portfolio
plt.grid(True)
plt.xlabel('expected volatility')
plt.ylabel('expected return')
plt.colorbar(label='Sharpe ratio')
```
